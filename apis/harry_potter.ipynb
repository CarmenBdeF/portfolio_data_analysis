```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://hp-api.onrender.com/api/characters"
```

```python
res = requests.get(url)
```

```python
res.status_code
res.reason
print(res.status_code)
print(res.reason)
```

```python
personajes = res.json()
personaje = personajes[0]
print(personaje['name'])
print(personaje['patronus'])
print(personaje['wand']['wood'])
```

```python
personajes = res.json()
personaje = personajes[1]
print(personaje['name'])
print(personaje['patronus'])
print(personaje['wand']['wood'])
```

```python
personajes = res.json()
personaje = personajes[2]
print(personaje['name'])
print(personaje['patronus'])
print(personaje['wand']['wood'])
```

```python
for personaje in personajes:
    print(personaje['name'])
```

```python
gryffindor = [p for p in personajes if p['house'] == 'Gryffindor']

for p in gryffindor:
    print(p['name'])
```

```python
for p in gryffindor:
    print(f"{p['name']} interpretado por {p['actor']}")
```

```python
personaje = personajes[0]
image_url = personaje["image"]
```

```python
for personaje in personajes[:3]:
    image_url = personaje["image"]
    if image_url:
        res = requests.get(image_url)
        if res.status_code == 200:
            image_data = res.content
            image = Image.open(BytesIO(image_data))

            plt.imshow(image)
            plt.axis("off")
            plt.title(personaje["name"])
            plt.show()
```
